```python
import pandas as pd

def clean_steam_data(file_path):
    print("--- Starting Data Cleaning Process ---")
    
    # 1. Load the raw dataset
    try:
        df = pd.read_csv(file_path)
        print(f"Successfully loaded dataset. Original Shape: {df.shape}")
    except FileNotFoundError:
        print("Error: steam.csv file not found. Please ensure it's in the same directory.")
        return

    # 2. Handling Missing Values
    # As identified, there are a few missing values in 'developer' and 'publisher' (approx 14 rows)
    missing_before = df.isnull().sum().sum()
    df.dropna(subset=['developer', 'publisher'], inplace=True)
    print(f"Dropped rows with missing developers/publishers. Total missing values resolved: {missing_before}")

    # 3. Data Type Standardization
    # Convert 'release_date' from string to actual Datetime object
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    print("Converted 'release_date' to datetime format.")

    # 4. Text Cleaning / Standardization
    # Ensure all categorical columns are properly capitalized (e.g., windows;mac;linux -> Windows;Mac;Linux)
    df['platforms'] = df['platforms'].str.title()
    df['categories'] = df['categories'].str.title()
    df['genres'] = df['genres'].str.title()
    print("Standardized text casing for platforms, categories, and genres.")

    # 5. Removing Potential Duplicates (Sanity Check)
    duplicates = df.duplicated(subset=['appid']).sum()
    if duplicates > 0:
        df.drop_duplicates(subset=['appid'], inplace=True)
        print(f"Removed {duplicates} duplicate game entries.")
    else:
        print("No duplicate entries found.")

    # 6. Export the Cleaned Dataset
    cleaned_filename = "steam_cleaned.csv"
    df.to_csv(cleaned_filename, index=False)
    print(f"--- Data cleaning complete! Cleaned dataset saved as '{cleaned_filename}' ---")
    print(f"Final Shape: {df.shape}")

# Execute the function
if __name__ == "__main__":
    # Ensure your raw dataset is named 'steam.csv' and is in the same folder as this script
    clean_steam_data('steam.csv')
```
